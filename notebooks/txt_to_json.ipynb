```python
import pandas as pd

# Replace 'file_path.txt' with your file path
data = pd.read_csv('data/inci_name.txt', delimiter = '\t') # if the file is tab-delimited
```

```python
data
```

```python
# Extract the 'INCI' column
column = data['INCI']

# Convert the column to a list
column_list = column.tolist()

# Convert the list to a Series
column_series = pd.Series(column_list)

column_series
```

**Note:** Unfortunately i've only noticied that we got lot's of "Fragrance" repeating itself only after exporting our JSON and running the script for the first time, need to see why that's happening!

```python
# Count the occurrences of each value
value_counts = column_series.value_counts()

# Print the count for "Fragrance"
print(value_counts["Fragrance"])
```

```python
filtered_df = data[data['INCI'] == 'Fragrance']

filtered_df
```

That's why. So what i'll need to do later on is to save all the values of product afther the ":" in another dataframe, then drop all the rows with value "Fragrance" in INCI, then append with the initial result to get a better value of names and then run again the program with this one.

```python
# Split the 'Product' column at ':', expand=True splits the string into different columns
split_df = filtered_df['PRODUCT'].str.split(':', expand=True)

split_df
```

```python
# Select the second column (index 1) which is the part after ':'
filtered_df['Product'] = split_df[1].str.strip()  # str.strip() is used to remove leading and trailing whitespace

filtered_df
```

**Note:** Now that we have only the Product name, let's get those and put into a list like the first part.

```python
# Extract the 'INCI' column
column2 = filtered_df['Product']

# Convert the column to a list
column_list2 = column2.tolist()

# Convert the list to a Series
column_series2 = pd.Series(column_list2)

column_series2
```

**Note:** Now i need to remove all the Fragance values from the first column_series and then add the column_series2

```python
# Drop all the values "Fragrance" from column_series
column_series = column_series[column_series != 'Fragrance']

column_series
```

```python
# Append column_series2 to column_series, not append, it's concact()
# Concatenate column_series and column_series2
column_series = pd.concat([column_series, column_series2], ignore_index=True)

column_series
```

**Note:** Now it's to export this series and then run the initial.py again

```python
# Export the Series to a JSON file
# Replace 'file_path.json' with your file path
column_series.to_json('data/INCI_names.json', orient='records')
```

**Note:** initial.py finished running with 2816 requests and now i want to analyze the results in here to see what i can expect after the extraction

```python
import json

# Load the JSON file
with open('data/INCI_results.json', 'r') as f:
    data = json.load(f)

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)

# Create a new DataFrame with just the 'name' and 'url' columns
df_name_url = df[['name', 'url']]

df_name_url
```
